```python

devElements = json.loads(requests.get("http://neo/api/elements").text)
standardElements = json.loads(requests.get("http://dispding/api/elements").text)
bulbElements = json.loads(requests.get("http://bulb02/api/elements").text)
plugElements = json.loads(requests.get("http://plug05/api/elements").text)
```

```python
elemSet = dict()
collectionSet = set()

def addElementCollection(collectionName, aList):
  collectionSet.add(collectionName)
  for e in aList:
    if e in elemSet:
      elemSet[e] |= {collectionName}
    else:
      elemSet[e] = {collectionName}

  return
```

```python
addElementCollection('CORE', ['ota', 'device'])
addElementCollection('SYSTEM', ['ntptime', 'time'])
addElementCollection('SYSTEM-FULL', ['state', 'ssdp'])
addElementCollection('CORE', ['alarm', 'analog', 'and', 'button', 'digitalin', 'digitalout', 'digitalsignal', 'log', 'map', 'or', 'pwmout', 'reference', 'remote', 'scene', 'schedule', 'switch', 'timer', 'value']) 
addElementCollection('DISPLAY', ['displaydot', 'displayline', 'displaytext', 'menu'])

addElementCollection('dev', devElements)
addElementCollection('standard', standardElements)
addElementCollection('bulb', bulbElements)
addElementCollection('plug', plugElements)

# elemSet
# collectionSet
```

```python
# Transform to a table (list of list)

theTable = []
# theTable.field_names = ['Element', 'plug', 'bulb', 'dev', 'CORE']
# cList = sorted(collectionSet)

for e in elemSet:
  row = [e]
  for c in collectionSet:
    if c in elemSet[e]:
      row.append(c)
    else:
      row.append('')
  theTable.append(row)
```
